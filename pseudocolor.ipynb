```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Plot z^2 = x^2 + y^2, a unit circle on 2D plane
# Intervals define the resolution of the final graph
def plot_mesh_data(axes, intervals):
    spacing = np.linspace(-1, 1, intervals)
    x, y = np.meshgrid(spacing, spacing)
    z = np.sqrt(x ** 2 + y ** 2)
    axes.pcolor(x, y, z)

# The subplots share both row and column
# from top to bottom & left to right, the resolution of graph is getting better
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex='col', sharey='row')
plot_mesh_data(ax1, 5)
plot_mesh_data(ax2, 10)
plot_mesh_data(ax3, 20)
plot_mesh_data(ax4, 100)
plt.show()
```

```python
# Different approach for subplotting
def plot_mesh_data(intervals, rows, cols, plot_num):
    spacing = np.linspace(-1, 1, intervals)
    x, y = np.meshgrid(spacing, spacing)
    z = np.sqrt(x ** 2 + y ** 2)
    plt.subplot(rows, cols, plot_num)
    plt.pcolor(x, y, z)

plot_mesh_data(5, 2, 2, 1)
plot_mesh_data(10, 2, 2, 2)
plot_mesh_data(20, 2, 2, 3)
plot_mesh_data(100, 2, 2, 4)
plt.show()
```
