## Load Packages and initialize style

```python
from sklearn import datasets
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('ggplot')
```

## Load built-in Iris dataset

```python
iris = datasets.load_iris()
iris.keys()
```

## Create numpy arrays and dataframe

```python
x = iris.data
y = iris.target
df = pd.DataFrame(x, columns=iris.feature_names)
```

## Plot Scatter Matrix

```python
pd.plotting.scatter_matrix(df, c=y, figsize=[8, 8], s=20, marker='D')
plt.show()
```
