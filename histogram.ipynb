```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import dataframe from csv
df = pd.read_csv(r"C:\Users\Chia Wei\Dropbox\Dataset\fifa.csv", delimiter=',')
df.head()
```

```python
# Trim columns

def trim_columns(df):
    old_columns = list(df.columns)
    new_columns = [col.strip() for col in list(df.columns)]
    columns_mapping = dict(zip(old_columns, new_columns))
    return df.rename(columns=columns_mapping)

print("Original columns: ")
print(df.columns)
df = trim_columns(df)
print("New columns: ")
print(df.columns)
```

```python
# Draw histogram
df.hist(column='height',bins=5)
plt.show()
```
