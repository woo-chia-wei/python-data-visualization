```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Plot z^2 = x^2 + y^2, a unit circle on 2D plane
# Intervals define the clarity of the final graph
def plot_mesh_data(axes, intervals):
    spacing = np.linspace(-1, 1, intervals)
    x, y = np.meshgrid(spacing, spacing)
    z = np.sqrt(x ** 2 + y ** 2)
    axes.pcolor(x, y, z)

# The subplots share both row and column
# from top to bottom & left to right, the clarify of graph is getting better
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex='col', sharey='row')
plot_mesh_data(ax1, 5)
plot_mesh_data(ax2, 10)
plot_mesh_data(ax3, 20)
plot_mesh_data(ax4, 100)
plt.show()
```
