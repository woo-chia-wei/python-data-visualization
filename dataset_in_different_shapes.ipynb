## Load packages

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

## XOR Dataset

```python
# Configuration
size = 500

# Create 2D-coordinates
X = np.vstack((np.random.rand(size, 2), 
               np.random.rand(size, 2) + (-1, -1), 
               np.random.rand(size, 2) + (-1, 0), 
               np.random.rand(size, 2) + (0, -1)))

# Create labels
Y = np.concatenate((np.repeat(0, size * 2),
               np.repeat(1, size * 2)))

# Combine X and Y to be training set with shuffled order
training = np.column_stack((X, Y))
np.random.shuffle(training)

# Create scatterplot
colors = ['blue', 'red']
plt.scatter(training[:, 0], training[:, 1], c=training[:, 2], alpha=0.5, cmap=ListedColormap(colors))
plt.axis('equal')
plt.show()

# Print first 10 records
print(training[:10])
```

## Donut Dataset

```python
# Configurations
size = 400
delta = 1

# Create 2D-coordinates
def donut_data(size, radius, delta):
    radii = np.interp(np.random.rand(size), [0, 1], [radius-delta, radius+delta])
    angles = np.interp(np.random.rand(size), [0, 1], [-np.pi, np.pi])
    return np.column_stack([radii * np.cos(angles), radii * np.sin(angles)]) 

X = np.vstack((donut_data(size, 10, 1),
               donut_data(size, 20, 1)))

# Create labels
Y = np.concatenate((np.repeat(0, size),
                    np.repeat(1, size)))

# Combine X and Y to be training set with shuffled order
training = np.column_stack((X, Y))
np.random.shuffle(training)

# Create scatterplot
colors = ['blue', 'red']
plt.scatter(training[:, 0], training[:, 1], c=training[:, 2], alpha=0.5, cmap=ListedColormap(colors))
plt.axis('equal')
plt.show()

# Print first 10 records
print(training[:10])
```
