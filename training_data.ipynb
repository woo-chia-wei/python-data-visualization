## Load packages

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

## XOR Dataset

```python
size = 500

X = np.vstack((np.random.rand(size, 2), 
               np.random.rand(size, 2) + (-1, -1), 
               np.random.rand(size, 2) + (-1, 0), 
               np.random.rand(size, 2) + (0, -1)))
Y = np.vstack((np.repeat(0, size * 2),
               np.repeat(1, size * 2)))
colors = ['blue', 'red']

plt.scatter(X[:, 0], X[:, 1], c=Y, alpha=0.5, cmap=ListedColormap(colors))
plt.show()
```
